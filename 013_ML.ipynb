```python
dataset=[['Milk','Egg','Bread','Butter'],
         ['Fish','Onion','Beans','Egg'],
         ['Milk','Bread','Butter','Egg'],
         ['Milk','Tea','Coffee','Egg']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
tr=TransactionEncoder()
```

```python
fq=tr.fit(dataset).transform(dataset)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(fq)
df
```

```python
from numpy import min_scalar_type
fi=apriori(df,min_support=0.3,use_colnames=True)
```

```python
fi
```

```python
dataset=[['pen'],['pencil'],['marker'],['paperclip'],['milk'],['marker'],['egg'],['bread'],['milk'],['bread'],['butter'],['marker']]
```

```python
fq=tr.fit(dataset).transform(dataset)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(fq,columns=tr.columns_)
df
```

```python
fi=apriori(df,min_support=0.1,use_colnames=True)
fi
```
